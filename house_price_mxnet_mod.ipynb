```python
import pandas as pd

import logging
logging.getLogger().setLevel(logging.DEBUG)
```

```python
train_data=pd.read_csv("kaggle/train.csv")
```

```python
all_features=train_data.iloc[:,1:]
```

```python
numeric_features=all_features.dtypes[all_features.dtypes != 'object'].index
```

```python
all_features[numeric_features] = all_features[numeric_features].apply(lambda x:(x-x.mean())/x.std())
```

```python
all_features = all_features.fillna(all_features.mean())
```

```python
data = pd.get_dummies(all_features.iloc[:,:79],dummy_na=True)
```

```python
label = all_features.iloc[:,79:]
```

```python
from mxnet import nd

train_features = nd.array(data.values)
train_labels = nd.array(label.values)
```

```python
import mxnet as mx
'''
label = mx.symbol.Variable('lin_reg_label')
net = mx.sym.Variable('data')
#net = mx.sym.FullyConnected(net, name='fc1', num_hidden=64)
#net = mx.sym.Activation(net, name='relu1', act_type="relu")
net = mx.sym.FullyConnected(net, name='fc2', num_hidden=1)
net = mx.sym.LinearRegressionOutput(net, label=label, name='lin_re')
mx.viz.plot_network(net, node_attrs={"shape":"oval","fixedsize":"false"})
'''

X = mx.sym.Variable('data')
Y = mx.symbol.Variable('lin_reg_label')
fc1  = mx.sym.FullyConnected(data=X, name='fc1', num_hidden = 150)
re1= mx.sym.Activation(fc1, name='relu1', act_type="relu")
fc2  = mx.sym.FullyConnected(data=re1, name='fc2', num_hidden = 1)
lro = mx.sym.LinearRegressionOutput(data=fc2, label=Y, name="lro")

mx.viz.plot_network(lro, node_attrs={"shape":"oval","fixedsize":"false"})
```

```python
split_train = int(len(train_features) * 0.8)

print(train_features[0:4])
batch_size = 200
train_iter = mx.io.NDArrayIter(train_features[:split_train, ], train_labels.reshape(-1,)[0:split_train], batch_size, shuffle=True,label_name='lin_reg_label')
val_iter = mx.io.NDArrayIter(train_features[split_train:,], train_labels.reshape(-1,)[split_train:], batch_size,label_name='lin_reg_label')
```

```python
mod = mx.mod.Module(
    symbol =lro ,
    data_names=['data'],
    label_names = ['lin_reg_label']# network structure
)
```

```python
'''

#medium level API
mod.bind(data_shapes=train_iter.provide_data, label_shapes=train_iter.provide_label)
# initialize parameters by uniform random numbers
#help(mod.init_params)
mod.init_params(force_init=True,initializer=mx.init.Uniform(scale=1))
#print(mod.get_params())
# use SGD with learning rate 0.1 to train
mod.init_optimizer(optimizer='sgd', optimizer_params=(('learning_rate', 0.1), ))
# use accuracy as the metric
metric = mx.metric.create('rmse')

#help(mod)
# train 5 epochs, i.e. going over the data iter one pass
for epoch in range(5):
    train_iter.reset()
    metric.reset()
    for batch in train_iter:
        #print(batch.data[0].asnumpy())
        #print(batch.label[0].asnumpy())
        #print(mod.get_params())
        mod.forward(batch, is_train=True)       # compute predictions
        mod.update_metric(metric, batch.label)  # accumulate prediction accuracy
        mod.backward()                          # compute gradients
        mod.update()                            # update parameters
    print('Epoch %d, Training %s' % (epoch, metric.get()))
'''
```

Note here: I encoutner a very strange problem that the mse increase dramatically with all the features included with small batch size. 

1. The problem is the model is too sensitive if only 1 sample in in batch. 
2. After increase the batch size, the problem has some relieve, but the validate mse increase along with the epochs. Its overfit issue, add more layers to the net, now its working well

```python

mod.fit(train_iter,
        eval_data=val_iter,
        optimizer='sgd',
        optimizer_params={'learning_rate':0.1},
        eval_metric='mse',
        batch_end_callback = mx.callback.Speedometer(batch_size, 5),
        num_epoch=10)
```
